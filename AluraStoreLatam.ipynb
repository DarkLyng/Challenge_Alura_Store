###**Introducción**

Se realizará un análisis de ventas y observará el rendimiento de cada una de las 4 tiendas para finalmente decidir cual tienda se venderá, para que el cliente pueda invertir en un negocio nuevo. La evaluación se hará basándose en estos cinco aspectos: facturación total de cada tienda (cuál tienda vende más); cuáles son las categorías más populares de cada tienda (cuáles productos se venden más en cada tienda); cuál es el promedio de evaluación de cada uno de los clientes; cuáles son los productos más y menos vendidos en cada tienda; y cuál es el costo promedio del envío.

### **Importación de datos**



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#**6. Gráficos**

Se obtienen diversos gráficos, utilizando barras y columnas según sus datos.

###**Gráfica 1: Ingresos por cada tienda**

```python
ingresos_tiendas = [t1_total_precio, t2_total_precio, t3_total_precio, t4_total_precio]
nombres_tiendas = ["Tienda 1", "Tienda 2", "Tienda 3", "Tienda 4"]

grafica_ingresos_tot = pd.DataFrame({"Tienda": nombres_tiendas, "Ingresos Totales": ingresos_tiendas})

grafica_ingresos_tot.plot(kind="bar", x="Tienda", y="Ingresos Totales", figsize=(8, 8), color=["green", "blue", "orange", "red"], legend= False,
                          xlabel= "Tiendas", ylabel= "Ingresos Totales", logy=True)
```

###**Gráfica 2: Distribución de categoría de productos por tienda**

```python
ventas_por_categoria = pd.DataFrame({"Tienda 1": t1_ventas_categ, "Tienda 2": t2_ventas_categ,"Tienda 3": t3_ventas_categ,
                                     "Tienda 4": t4_ventas_categ}).fillna(0)

ventas_por_categoria.plot(kind="barh", figsize=(15, 8), xlabel="Categoría del Producto", ylabel="Cantidad de Ventas")
```

###**Gráfica 3: Opiniones de clientes por tienda**

```python

calif_tiendas = [t1_calificacion_prom, t2_calificacion_prom, t3_calificacion_prom, t4_calificacion_prom]

grafica_calif_clientes = pd.DataFrame({"Tienda": nombres_tiendas, "Calificación": calif_tiendas})
grafica_calif_clientes.plot(kind="bar", x="Tienda", y="Calificación", figsize=(8, 5), color=["green", "blue", "orange", "red"], legend= False,
                          xlabel= "Tiendas", ylabel= "Calificación", logy=True)
```

###**Gráfica 4: Costos de envío promedio por tienda**

```python
envio_tiendas = [t1_prom_envio, t2_prom_envio, t3_prom_envio, t4_prom_envio]

grafica_envio_prom = pd.DataFrame({"Tienda": nombres_tiendas, "Costo de envío": envio_tiendas})
grafica_envio_prom.plot(kind="bar", x="Tienda", y="Costo de envío", figsize=(8, 5), color=["green", "blue", "orange", "red"], legend= False,
                          xlabel= "Tiendas", ylabel= "Costo de envío", logy=True)
```
